```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("energydata_complete.csv")
df.head()
```

```python
df.info()
```

```python
desc = df.describe()
print(desc)
```

```python
plt.boxplot(df["Appliances"])
plt.title("Boxplot - Konsumi i energjise (Appliances)")
plt.ylabel("Wh")
plt.show()
```

```python
main_vars = [
    "Appliances", "lights",
    "T1","T2","T3","T4","T5","T6","T7","T8","T9",
    "RH_1","RH_2","RH_3","RH_4","RH_5","RH_6","RH_7","RH_8","RH_9",
    "T_out","RH_out","Windspeed","Visibility","Press_mm_hg"
]
desc_stats = df[main_vars].agg(['mean', 'median', 'std', 'min', 'max']
                              )
desc_stats
```

```python
stats_mean = df [[
    "Appliances",
    "T1",
    "RH_1",
    "lights",
    "T_out"
]].mean()
    
stats_mean.plot(kind='bar', figsize=(8,5), color="#e090d8", edgecolor="black")
plt.title("Mesatarja e variablave kryesore të sistemit HVAC")
plt.ylabel("Vlera mesatare")
plt.xlabel("Variablat")
plt.show()

```

```python
plt.hist(df["Appliances"],bins=30,color='#c2275a',edgecolor="black")
plt.title("Histogrami i konsumit te energjise (Appliances)")
plt.xlabel("Konsumi (Wh)")
plt.ylabel("Frekuenca")
plt.show()
```

```python
plt.hist(df["T1"],bins=30, color="#6e917f", edgecolor="black")
plt.title("Histogrami i temperatures se brendshme (T1)")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Frekuenca")
plt.show()
```

```python
energy_per_temp = df["Appliances"] / df["T1"]
df["energy_per_temp"] = energy_per_temp
df[["Appliances", "T1", "energy_per_temp"]].head()
# Sa energji elektrike shpenzohet për një njësi temperature

plt.figure(figsize=(7,5))
plt.hist(df["energy_per_temp"], bins=40, color="#6f96e3",edgecolor="black")
plt.xlabel("Energji/Temperature")
plt.ylabel("Frekuenca")
plt.title("Shperndarja e raportit Energji-Temperature")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["T1"],df["Appliances"], alpha=0.4, color="#6789a6")
plt.title("Temperatura e brendshme (T1) vs Konsumi i energjise (Appliances)")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Konsumi i energjise (Wh)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["T_out"],df["Appliances"],alpha=0.4,color="#9271a3")
plt.title("Temperatura e jashtme (T_out) vs Konsumi i energjise (Appliances)")
plt.xlabel("Temperatura e jashtme (°C)")
plt.ylabel("Konsumi i energjise (Wh)")
plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.plot(df["T1"][:300], label="T1 – temperatura e brendshme")
plt.plot(df["T_out"][:300], label="T_out – temperatura e jashtme")
plt.xlabel("Koha")
plt.ylabel("Temperatura (°C)")
plt.title("Krahasimi i temperaturës së brendshme dhe të jashtme")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
top_days=df.nlargest(10,"Appliances")[["date","Appliances"]].set_index("date")
top_days.plot(kind="bar", figsize=(10,5), color="#d48d4e", edgecolor="black", title="10 Ditet me Konsumin me te Larte te Energjise")
plt.ylabel("Konsumi i energjisë (Wh)")
plt.xlabel("Ditet")
plt.show()
```

```python
import seaborn as sns
corr_vars=["Appliances","T1","RH_1","lights","T_out"]
corr_matrix=df[corr_vars].corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Matrica e korrelacionit - variablat kryesore")
plt.show()
```
